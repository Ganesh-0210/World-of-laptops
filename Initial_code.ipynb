```python
Lap_Name = []
Lap_Price = []
Processor = []
```

```python
url = "https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D75000&page=1"
r = requests.get(url)

soup = BeautifulSoup(r.text, "html")

names = soup.find_all("div", class_ = "_4rR01T")

for i in names:
    name = i.text
    Lap_Name.append(name)
print(Lap_Name)
print(len(Lap_Name))
```

```python
prices = soup.find_all("div",class_ = "_30jeq3 _1_WHN1")
for i in prices:
    price = i.text
    Lap_Price.append(price)
print(Lap_Price)
print(len(Lap_Price))
```

```python
specs = soup.find_all("div",class_ = "fMghEO")
for i in specs:
    spec = i.text
    Specifications.append(spec)
print(Specifications)
print(len(Specifications))
```

```python
for i in range(2,27):
    url = "https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D75000&page=" + str(i)
    r = requests.get(url)

    soup = BeautifulSoup(r.text, "lxml")

    names = soup.find_all("div", class_ = "_4rR01T")

    for i in names:
        name = i.text
        Lap_Name.append(name)
        
    prices = soup.find_all("div",class_ = "_30jeq3 _1_WHN1")
    
    for i in prices:
        price = i.text
        Lap_Price.append(price)
        
    specs = soup.find_all("div",class_ = "fMghEO")
    for i in specs:
        spec = i.text
        Specifications.append(spec)
        
    
               
print(len(Lap_Name))
print(len(Lap_Price))
print(len(Specifications))
```

```python
Lap_Name.clear()
Lap_Price.clear()
Specifications.clear()
```

```python
A = {"Lap_Name" : Lap_Name, "Lap_Price" : Lap_Price, "Specifications" : Specifications}
print(A)
```
