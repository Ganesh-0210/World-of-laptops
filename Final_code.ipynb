```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
Processor = []
Lap_Name = []
Lap_Price = []
```

```python
url = "https://www.flipkart.com/computers/laptops/pr?sid=6bo%2Cb5g&q=laptops+under+60000&otracker=categorytree&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D60000&sort=relevance&page=1"
r = requests.get(url)
soup = BeautifulSoup(r.text, "html")
print(soup)
```

```python
next_page = soup.find("a", class_ = "_1LKTO3").get("href")
print(next_page)
```

```python
cnp = "https://www.flipkart.com" + next_page  # complete link of next page
print(cnp)
```

```python
for i in range(2,10):
    url = "https://www.flipkart.com/computers/laptops/pr?sid=6bo%2Cb5g&q=laptops+under+60000&otracker=categorytree&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D60000&sort=relevance&page=" + str(i)
    r = requests.get(url)
    # print(r)

    soup = BeautifulSoup(r.text, "html")
    # print(soup)    
    
    next_page = soup.find("a", class_ = "_1LKTO3").get("href")
    cnp = "https://www.flipkart.com" + next_page  # complete link of next page
    print(cnp)
```

```python
for i in range(2,10) :
    
    url = "https://www.flipkart.com/computers/laptops/pr?sid=6bo%2Cb5g&q=laptops+under+60000&otracker=categorytree&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D60000&sort=relevance&page=" + str(i)

    r = requests.get(url)

    soup = BeautifulSoup(r.text, "html")
    box = soup.find_all('li')
    box_titles = [title.text for title in box]
    for i in box_titles:
        if (('Intel Core' in i) | ('AMD Ryzen' in i) | ('MediaTek' in i) | ('Dual Core' in i) | ('Quad Core' in i) | ('Pentium' in i) | ('Snapdragon' in i) | ('Zen' in i)):
            Processor.append(i)
          
    names = soup.find_all("div", class_ = "_4rR01T")

    for i in names:
        name = i.text
        Lap_Name.append(name)
        
    prices = soup.find_all("div",class_ = "_30jeq3 _1_WHN1")
    
    for i in prices:
        price = i.text
        Lap_Price.append(price)
        
#print(box)
#print(box_titles)
#print(Processor)
print(len(Processor))
print(len(Lap_Name))
print(len(Lap_Price))
```

```python
Processor.clear()
Lap_Name.clear()
Lap_Price.clear()
```
